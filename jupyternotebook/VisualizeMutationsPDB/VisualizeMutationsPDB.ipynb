```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import rdMolTransforms as rdmt
import numpy as np
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem.Draw import MolDrawing, DrawingOptions
from rdkit.Geometry import rdGeometry as geom
import py3Dmol
from rdkit import Chem
from rdkit.Chem import AllChem
from ipywidgets import interact, interactive, fixed
from ipywidgets import interact, widgets
from IPython.display import display
import py3Dmol
```

```python
def get_mutations(mutationfile):
    muts = {}
    with open(mutationfile,'r') as f:
        for line in f:
            key = line.strip()
            muts[key] = key[1:-1]
    return muts
```

```python
def set_view_to_structure(view,dictionary_w_mutations, color='cyan'):
    for i in dictionary_w_mutations.keys():
        mod_res = {'resi': dictionary_w_mutations[i], 'chain': 'A'}
        col = color
        c_col = col + 'Carbon'
        view.addStyle(mod_res, {'stick':{'colorscheme':c_col, 'radius': 0.2}})
        view.addLabel(i, {'fontSize':10,'fontColor': 'black','backgroundColor':'white'}, {'resi': dictionary_w_mutations[i], 'chain': 'A'})
```

```python
def intersection(lst1, lst2): 
    lst3 = [value for value in lst1 if value in lst2] 
    return lst3
```

```python
def get_unique_mutations(common_mutations, dictionary_w_mutations):
    list(map(dictionary_w_mutations.__delitem__, filter(dictionary_w_mutations.__contains__,common_mutations)))
    return dictionary_w_mutations
```

```python
PDBFILE = "1.pdb"
MUTFILE = "pssm_muts.txt"
MUTFILE2 = "gremlin_muts.txt"
```

```python
pdbdata=open(PDBFILE, 'r').read()
view=py3Dmol.view()
proteinchain = {'chain':'A'}
view.addModel(pdbdata,'pdb')
view.setStyle(proteinchain,{'cartoon': {'color':'white'}})
design_residues_nr = get_mutations(MUTFILE)
design_residues_nr2 = get_mutations(MUTFILE2)
```

```python
# union between the mutational files
common_mutations = intersection(design_residues_nr.keys(),design_residues_nr2.keys() )
design_residues_nr = get_unique_mutations(common_mutations,design_residues_nr)
design_residues_nr2 = get_unique_mutations(common_mutations,design_residues_nr2)
cmon_mutations = {}
for i in common_mutations:
    cmon_mutations[i] = i[1:-1]
```

```python
set_view_to_structure(view, design_residues_nr,'cyan')
set_view_to_structure(view, design_residues_nr2,'red')
set_view_to_structure(view, cmon_mutations,'green')
view.show()
```

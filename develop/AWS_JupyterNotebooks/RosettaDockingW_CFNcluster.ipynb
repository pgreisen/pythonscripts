```python
def get_availability_zones():
    import boto3
    ec2 = boto3.client('ec2')
    # Retrieves all regions/endpoints that work with EC2
    response = ec2.describe_regions()
    ## print 'Regions:', response['Regions'] 
    # Retrieves availability zones only for region of the ec2 object
    response = ec2.describe_availability_zones()
    return response['AvailabilityZones']
##get_availability_zones()
```

```python
def get_spot_price(instance_type='m3.medium',regionname='us-east-1',zonename='us-east-1a'):
    print "Region requested: ", regionname
    import boto3
    client=boto3.client('ec2',region_name=regionname)
    prices =client.describe_spot_price_history(InstanceTypes=[instance_type],MaxResults=1\
                                               ,ProductDescriptions=['Linux/UNIX (Amazon VPC)'],AvailabilityZone=zonename)
    return prices['SpotPriceHistory']# [0]
```

```python
# Get IP address of EC2 instances
from collections import defaultdict

import boto3

"""
A tool for retrieving basic information from the running EC2 instances.
"""

# Connect to EC2
ec2 = boto3.resource('ec2')

# Get information for all running instances
running_instances = ec2.instances.filter(Filters=[{
    'Name': 'instance-state-name',
    'Values': ['running']}])

print running_instances

ec2info = defaultdict()
for instance in running_instances:
    for tag in instance.tags:
        if 'Name'in tag['Key']:
            name = tag['Value']
    # Add instance info to a dictionary         
    ec2info[instance.id] = {
        'Name': name,
        'Type': instance.instance_type,
        'State': instance.state['Name'],
        'Private IP': instance.private_ip_address,
        'Public IP': instance.public_ip_address,
        'Launch Time': instance.launch_time
        }

attributes = ['Name', 'Type', 'State', 'Private IP', 'Public IP', 'Launch Time']
for instance_id, instance in ec2info.items():
    for key in attributes:
        print("{0}: {1}".format(key, instance[key]))
    print("------")
```

```python
# AWS run script for docking
def setup_docking_w_rst(parm="LG.params",pdbfile="Csa1G044890_1_aligned.pdb",prefix="dock",runs="10"):
    template='''
#!/bin/sh -f
# update machine
sudo yum -y update;
sudo yum install -y aws-cli;
cd /home/ec2-user;

# add credentials to account
# we do this by adding the information to a file
# ~/.aws/config that looks like this:
mkdir ~/.aws/;
cp config ~/.aws/;
tar zxf files.tgz;
rm files.tgz
# parameter needed to be set
region=eu-central-1;
bucketname=RosettaExecutables_20160925;

# get the right files from S3
aws s3 cp s3://RosettaExecutables_20160925/database.tgz . --region $region;
aws s3 cp s3://$bucketname/rosetta_scripts.static.linuxgccrelease . --region $region;
chmod +x ./*;
tar zxf database.tgz;
rm database.tgz;
# parameters defined
exe=./rosetta_scripts.static.linuxgccrelease;
chmod a+x $exe;
database=./database;
xml=./dock.xml;
params='''+parm+''';
pdbfile='''+pdbfile+''';
output='''+dock+''';
runs='''+runs+''';
# run program
for i in $(seq 1 $runs);
do
$exe -database $database  -parser:protocol $xml -in:file:s $pdbfile -out:prefix $output -in:file:native $pdbfile @flags -extra_res_fa $params > /dev/null & echo "Done";
done
'''
```

```python
def gen_qsub_file(path="~",nstruct=500,prefix="",cstfile="constraintfile.cst",*args):
    extra_fa = ""
    for i in list(args):
        extra_fa += i+" "
    template='''#!/bin/sh -f
# argument 1 is the pdbfile
# argument 2 is the parameter file with conf at the end
path='''+path+'''
database='''+path+'''/database;
exe='''+path+''';
xml=dock.xml;
prefix='''+prefix+''';
nstruct='''+str(nstruct)+''';
$exe/rosetta_scripts.static.linuxgccrelease -database $database \
-parser:protocol $xml -in:file:s $pdbfile -in:file:native $pdbfile \
-out:prefix $prefix -nstruct $nstruct -out:file:scorefile $prefix\_dock.sc \
-packing:ex1 -packing:ex2 cstfile='''+cstfile+''' -extra_res_fa '''+extra_fa+''' @flags
    '''
    return template
```
